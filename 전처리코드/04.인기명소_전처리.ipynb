# 1. 라이브러리 불러오기

```python
import pandas as pd
import numpy as np
```

# 2. 데이터 불러오기

```python
df_hotspot_2019 = pd.read_csv('D:/04.국립공원자료/2019_hotspot.csv')
df_hotspot_2020 = pd.read_csv('D:/04.국립공원자료/2020_hotspot.csv')
df_model_final = pd.read_csv('D:/04.국립공원자료/df_model_final.csv')
```

# 3. 2019_hotspot & 2020_hotspot 합치기 (35만 개 -> 1 데이터프레임)

```python
df_hotspot_2019.drop('Unnamed: 0',axis=1, inplace=True)
df_hotspot_2019
```

```python
df_hotspot_2020.drop('Unnamed: 0',axis=1, inplace=True)
df_hotspot_2020
```

```python
hotspot_final = pd.concat([df_hotspot_2019, df_hotspot_2020], ignore_index=True)
hotspot_final.drop('Unnamed: 0',axis=1, inplace=True)
hotspot_final
```

```python
hotspot_final.rename(columns={'f_Name':'f_name'}, inplace=True)
hotspot_final
```

# 4. 이전 모델링용 데이터 프레임에 hotspot_final 합치기

```python
df_model_final.drop('Unnamed: 0',axis=1, inplace=True)
df_model_final
```

```python
df_model_real_final = pd.merge(df_model_final, hotspot_final, how ='left', on='f_name')
df_model_real_final
```

# 5. hospot feature -> one-hot encoding

```python
n_list=['자운봉','백운대','인수봉','비봉','만경대',
        '향로봉','만장봉','족두리봉','선인봉', '오봉',
        '송추계곡','사패산/포대능선','사모바위','의상능선','숨은벽',
        '대동문','북한산성','진관사','진흥왕순수비','대남문']

for n in n_list:
    df_model_real_final[n].replace('Y',1, inplace = True)
    df_model_real_final[n].replace('N',0, inplace = True)
```
