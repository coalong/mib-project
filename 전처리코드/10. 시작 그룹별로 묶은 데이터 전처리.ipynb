# 2019년 데이터 전처리

```python
import pandas as pd
import numpy as np
```

```python
# 그룹별로 묶어진 2019년 데이터 불러오기

Area_A_2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/2019_Area_A.csv')
Area_B_2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/2019_Area_B.csv')
Area_C_2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/2019_Area_C.csv')
Area_D_2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/2019_Area_D.csv')
Area_E_2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/2019_Area_E.csv')
Area_F_2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/2019_Area_F.csv')
Area_G_2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/2019_Area_G.csv')
Area_H_2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/2019_Area_H.csv')
```

```python
# 불러온 데이터의 전체 개수 확인

print(len(Area_A_2019) + len(Area_B_2019) + len(Area_C_2019) + len(Area_D_2019) 
      + len(Area_E_2019) + len(Area_F_2019) + len(Area_G_2019) + len(Area_H_2019))  
```

```python
# Area_A ~ Area_H 까지의 데이터프레임을 하나의 데이터프레임으로 만들기

df2019 = pd.concat([Area_A_2019, Area_B_2019, Area_C_2019, 
                    Area_D_2019, Area_E_2019, Area_F_2019, Area_G_2019, Area_H_2019]) # 파일 합치기
df2019.drop('id', axis=1, inplace=True) #'id' 열 drop
df2019.reset_index(inplace=True) # index 번호 재설정
df2019.drop('index', axis=1,inplace=True) # index 열 삭제
df2019.drop('field_1', axis=1,inplace=True) # filed_1 열 삭제
df2019
```

```python
# 통계 excel 파일 불러오기

xlsx2019 = pd.read_excel('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터/국립공원공단_북한산트레킹2019_2020_이용자연령성별데이터.xlsx', header=1)
xlsx2019 = xlsx2019[['번호', '소모Kcal','운동 종목', '성별', '연령','gpx파일']]
xlsx2019 
```

```python
# 열 이름 바꾸기
xlsx2019 = xlsx2019.rename({'gpx파일' : 'f_name',
                            '번호':'number', 
                            '소모Kcal': 'Kcal', 
                            '운동 종목':'sports_type',
                            '성별':'sex',
                            '연령':'age'}, axis=1)
xlsx2019
```

```python
# 2019년 통계파일과 범위별 csv파일 합치기
df2019_all = pd.merge(df2019,xlsx2019,on='f_name',how='outer')
df2019_all
```

```python
# 2019년 데이터  결측값이 있는 열 제거
df2019_all.dropna(inplace=True)

# 성별의 'NONE'을 결측값으로 바꾸기
df2019_all['sex'].replace('NONE',np.nan, inplace=True)

#결측값이 있는 열 제거
df2019_all.dropna(inplace=True)
df2019_all
```

```python
# groupby로 나이와 성별로 묶기
count_2019 = df2019_all.groupby(['age', 'sex']).count()
count_2019
```

```python
# top_speed의 이상치를 알아보기 위한 작업

# box plot의 이상치
Q1_speed = df2019_all['top_speed'].quantile(0.25)
Q3_speed = df2019_all['top_speed'].quantile(0.75)
IQR_speed = Q3_speed - Q1_speed

upper_bound_speed = Q3_speed + (IQR_speed*1.5) # 이상치 상한선
lower_bound_speed = Q1_speed - (IQR_speed*1.5) # 이상치 하한선
```
