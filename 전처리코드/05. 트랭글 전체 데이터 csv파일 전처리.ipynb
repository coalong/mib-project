### 결측치 있는 열 모두 drop하고 feature 데이터타입 바꾸기

2019년도 데이터 전처리

```python
outer_2019['성별'] = outer_2019['성별'].replace('NONE', np.nan)
df2019 = outer_2019.dropna()
df2019.reset_index(inplace=True)
df2019.drop(['index'], axis=1, inplace=True)
df2019
```

```python
df2019['성별'].unique()
```

```python
df2019.info()
```

```python
# object타입인 열의 unique 확인하기

print(df2019['운동 종목'].unique())
print(df2019['성별'].unique())
```

```python
df2019['운동 종목'] = df2019['운동 종목'].replace('HIKING', 0) #운동종목의 HIKING을 0으로 바꾸기
df2019['운동 종목'] = df2019['운동 종목'].replace('WALKING', 1) #운동종목의 WALKING을 1으로 바꾸기

df2019['성별'] = df2019['성별'].replace('MALE', 0) #남자는 0으로 바꾸기
df2019['성별'] = df2019['성별'].replace('FEMALE', 1) #여자는 1으로 바꾸기
```

```python
df2019.info()
```

```python
# csv 파일로 저장하기
df2019.to_csv('df2019_preprocessing.csv')
```
