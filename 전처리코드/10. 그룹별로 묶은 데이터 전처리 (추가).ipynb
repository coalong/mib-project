```python
import pandas as pd
import numpy as np
```

## 2019년

### 계절 정보 추가하기

```python
df2019 = pd.read_csv('C:/Users/박연재/Desktop/국립공원공단/트랭글데이터 분석/trangle2019_preprocess/df2019_preprocess_0819.csv')
df2019.drop('Unnamed: 0', axis=1, inplace=True)
df2019
```

```python
import time

start = time.time()  # 시작 시간 저장

season = [] #계절 정보를 추가할 리스트 만들기

for i in  df2019.index:
    month = int(df2019['p1_time'][i].split()[0].split('/')[1]) #'p1_time'에서 월 정보만 저장
    
    if (month == 3 or month == 4 or month == 5): # 3, 4, 5월은 봄
        season.append('봄')
    elif (month == 6 or month == 7 or month == 8): # 6, 7, 8월은 여름
        season.append('여름')
    elif (month == 9 or month == 10 or month == 11): # 9, 10, 11월은 가을
        season.append('가을')
    elif (month == 12 or month == 1 or month == 2): # 12, 1, 2월은 겨울
        season.append('겨울')

df2019['season'] = season

print("time :", time.time() - start)  # 현재시각 - 시작시간 = 실행 시간
```

```python
df2019
```

### 시간대 정보 추가하기

```python
time_div = [] #시간대 정보를 저장할 리스트 생성

for i in  df2019.index:
    t = int(df2019['p1_time'][i].split()[1].split(':')[0])
    
    if (t == 5 or t == 6 or t == 7 or t == 8 or t == 9 or t == 10): # 5 ~ 10시 : 아침
        time_div.append('아침')
    elif (t == 11 or t == 12 or t == 13 or t == 14 or t == 15 or t == 16): # 11 ~ 16시 : 점심
        time_div.append('점심')
    elif (t == 17 or t == 18 or t == 19 or t == 20): # 17 ~ 20시 : 저녁
        time_div.append('저녁')
    else: # 21 ~ 4시 :야간
        time_div.append('야간')

df2019['time_div'] = time_div
```
