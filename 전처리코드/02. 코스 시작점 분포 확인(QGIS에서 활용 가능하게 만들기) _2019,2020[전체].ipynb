## 1.필요한 라이브러리 호출

```python
# 라이브러리 호출

import gpxpy
import pandas as pd
import os
from glob import glob
import geopandas
from datetime import datetime
```

## 2. CSV파일 불러오기 -> pandas.DataFrame

```python
df1 = pd.read_csv('D:/Trangle_Data/Trangle_GPX/2019_done.csv')
df1.head()
```

```python
df2 = pd.read_csv('D:/Trangle_Data/Trangle_GPX/2020_done.csv')
df2.head()
```

## 3. pandas.DataFrame -> geopandas.GeoDataFrame 

```python
# df1에서 '파일명', '코스의 첫 포인트 경도&위도'를 GeoDataFrame으로 변경

gdf1 = geopandas.GeoDataFrame(
    df1[['f_name', 'p1_lat', 'p1_lon']],
    geometry=geopandas.points_from_xy(df1['p1_lat'], df1['p1_lon']))
gdf1.head()
```

```python
# df2에서 '파일명', '코스의 첫 포인트 경도&위도'를 GeoDataFrame으로 변경

gdf2 = geopandas.GeoDataFrame(
    df2[['f_name', 'p1_lat', 'p1_lon']],
    geometry=geopandas.points_from_xy(df2['p1_lat'], df2['p1_lon']))
gdf2.head()
```

## 4. geopandas.GeoDataFrame -> CSV 파일
* QGIS에서 CSV파일을 Shapefile로 분포 확인 가능

```python
# df1, df2를 csv 파일로 저장 

gdf1.to_csv('D:/Trangle_Data/Trangle_GPX/2019_geo_df/2019_geo_df.csv')
gdf2.to_csv('D:/Trangle_Data/Trangle_GPX/2020_geo_df/2020_geo_df.csv')
```
