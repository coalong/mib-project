```python
# 인코딩 된거 되돌리기

df2019['season'].replace(0, '봄', inplace=True) # 봄 : 0
df2019['season'].replace(1, '여름', inplace=True) # 여름 : 1
df2019['season'].replace(2, '가을', inplace=True) # 가을 : 2
df2019['season'].replace(3, '겨울', inplace=True) # 겨울 : 3

df2019['time_div'].replace(0, '아침', inplace=True) # 아침 : 0
df2019['time_div'].replace(1, '점심', inplace=True) # 점심 : 1
df2019['time_div'].replace(2, '저녁', inplace=True) # 저녁 : 2
df2019['time_div'].replace(3, '야간', inplace=True) # 야간 : 3

df2019['sports_type'].replace(0, 'HIKING', inplace=True) #HIKING : 0
df2019['sports_type'].replace(1, 'WALKING', inplace=True) #WALKING : 0

df2019['sex'].replace(0,'MALE', inplace=True) # MALE : 0
df2019['sex'].replace(1, 'FEMALE', inplace=True) # FEMALE : 0
```

### 2019 gpx 데이터

```python
# gpx 데이터에 해당하는 feature만 불러오기

df2019_gpx = df2019[['name', 'f_name', 'track_length', 'top_speed', 'lowest_ele', 'top_ele',
       'moving_time', 'stopped_time', 'elevation_gain', 'p1_lat', 'p1_log',
       'p1_time', 'dist/hour']]
df2019_gpx
```

```python
# 최고 고도가 900 이상인 행 제거

df2019_gpx.loc[df2019_gpx['top_ele'] >= 900]
df2019_gpx = df2019_gpx.loc[df2019_gpx['top_ele'] < 900]
```

```python
df2019_gpx
```

```python
# 등산하는데 걸린 총 시간

df2019_gpx['tracking_time'] = df2019_gpx['moving_time'] + df2019_gpx['stopped_time']
#df2019_gpx['mid_ele'] = (df2019_gpx['top_ele']+df2019_gpx['lowest_ele'])/2
df2019_gpx
```

```python
#df2019_gpx = df2019_gpx.drop('mid_ele', axis=1)
df2019_gpx
```

```python
# 인덱스 재설정하기

df2019_gpx.reset_index(inplace=True)
df2019_gpx.drop('index', axis=1, inplace=True)
df2019_gpx
```

```python
df2019_gpx.loc[df2019_gpx['top_ele'] >= 500].count()
```

```python
# 정상 여부 추가

top = []

for i in range(len(df2019_gpx)):
    ele = df2019_gpx['top_ele'][i]
    
    if(ele>700):
        top.append(1) # 정상 도달
    else:
        top.append(0) # 정상 미도달
        
df2019_gpx['reached_top'] = top
```

```python
df2019_gpx['reached_top'].value_counts()
```

```python
import math

# df2019_gpx['level'] = math.sqrt(df2019_gpx['track_length']*df2019_gpx['elevation_gain']*2)
# 코스의 난이도를 판별하기 위한 열 추가
# (트랙 전체 길이)*(누적 고도 합)*2 의 제곱근

level = []
for i in range(len(df2019_gpx)):
    level.append(math.sqrt(df2019_gpx['track_length'][i]*df2019_gpx['elevation_gain'][i]*2))

#level
```
